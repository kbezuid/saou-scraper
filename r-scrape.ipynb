```r
library(xml2)
library(rvest)
library(stringr)
library(magrittr)
library(jsonlite)
```

```r
base_url <- "http://www.saou.co.za/poste/jobs/"
current_url <- base_url
```

```r
sanitise <- function(nodes) {
    converted <- as.character( html_text(nodes) )
    converted <- gsub("\\s", "", converted) 
    # converted <- gsub("\n","",converted)
    # converted <- gsub("\t","", converted)
    return(converted)
}
```

```r
statuses <- c()
titles <- c()
subtitles <- c()
regions <- c()
types <- c()
```

```r
page <- read_html(base_url)
```

```r
number_of_pages <- html_nodes(page, '#all_jobs > div.dashboard-pagination > a:nth-child(5)') %>% html_text %>% as.numeric
```

```r
for( current_page in 1:number_of_pages) {
    message(sprintf("Scraping page %d of %d : %s",current_page, number_of_pages, current_url))
    all_jobs <- html_nodes(page, 'div#all_jobs')
    job_titles <- html_nodes(all_jobs, 'div.job-listing-name > h4')
    status_nodes <- html_nodes(job_titles, 'span')
    statuses <- c(statuses,sanitise(status_nodes))
    xml_remove(status_nodes)
    titles <- c(titles,sanitise(html_nodes(all_jobs, 'div.job-listing-name > h4')))
    subtitles <- c(subtitles,sanitise(html_nodes(all_jobs,'p.job-listing-summary')))
    regions <- c(regions,sanitise(html_nodes(all_jobs,'div.job-listing-region')))
    types <- c(types,sanitise(html_nodes(all_jobs,'div.job-listing-type')))
    
    current_url <- sprintf('%spage/%d',base_url,current_page+1)
    page <- read_html(current_url)
}
```

```r
jobs <- data.frame( Status=statuses, Title=titles, Subtitles=subtitles, Regions=regions)
```
